# Data Cleaning:
According to the data dictionary, several variables with NaN entries (Not a Number) truly were missing these features (e.g., no pool, no fence, no fireplaces, etc.). These data were imputed with na. There were a few minor inconsistencies in the training data set. For instance, one home had no garage square footage or any other values that might indicate it had a garage, yet it is listed as having a detached garage. In these cases, the most consistent option was imputed (i.e., the detached garage entry was deemed a data entry error).

On another note, 330 properties had no value for lot frontage. An examination of a map of Ames, IA revealed several properties that lack borders on streets (see https://ww2.amstat.org/publications/jse/v19n3/decock/AmesResidential.pdf). Therefore, I decided properties that lack data for lot frontage truly had no lot_frontage.

Finally, some houses have no garages. I decided to fill in a zero value for the year the garage was built, treating this variable like a categorical variable, not a continuous variable, so it is only important that this value can be distinguished from other values of this variable.

```python
train_df.drop(columns='pid', inplace=True)
```

```python
def find_NaNs_left_to_clean(df=train_df):
    return df.loc[:, df.isnull().sum() != 0].isnull().sum()
```

```python
find_NaNs_left_to_clean(train_df)
```

```python
feature_not_present = ['pool_qc', 'misc_feature', 'pool_qc', 'alley',
                       'fence', 'misc_feature', 'mas_vnr_type', 'bsmt_qual',
                       'bsmt_cond', 'bsmt_exposure', 'bsmtfin_type_1',
                       'bsmtfin_type_2', 'fireplace_qu', 'garage_type',
                       'garage_finish', 'garage_qual', 'garage_cond',]
```

```python
for feature in feature_not_present:
    train_df[feature].fillna('na', inplace=True)
```

# Engineer Variables:
So, discussing with my classmates, it seems that we are all following similar paths, trying to throw all of our variables into Lasso and count on it to decide what variables to push forward into our final model. Thus far, I have heard about two similar approaches, neither of which made models that fared better than our baseline "bad model." So, let's try a different approch:  engineer variables, look at interactions between variables and see if we can create a model that might be expected to fare better from a theoretical perspective.

Let's choose variables that are most likely to be maximally predictive:  overall_qual, gr_liv_area, 1st_fl_sf, year_built, full_bath, year_remodel/add, totrms_abvgr, and fireplaces. Let's feed these into a simple linear regression model,adding the interactions between overall_qual and gr_liv_area, between year_built and year remodeled, and between the total number of rooms above ground, the number of fireplaces, and the number of full bathrooms.

# Linear Regression:

To create a baseline model, we will use 

```python
# Create target variable, y, and df of predictor var's, X
y = train_df_dummies.saleprice
X = train_df_dummies.drop(columns='saleprice')
```

```python
lm = LinearRegression()
model = lm.fit(X, y)
```

```python
model.score(X, y)
```

An exceedingly high R^2. This model is almost certainly over-fit. To deal with over-fitting, we will separate the data into two sets, a "train set" and a "test set." We will then use the "train set" to train the model, keeping the "test set" to test our model as if it were encountering new, unseen data.

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
slr = LinearRegression()
slr.fit(X_train, y_train)
```

```python
slr.score(X_train, y_train)
```

```python
slr.score(X_test, y_test)
```

The train test split verifies that the linear regression model is overfit. In other words the model is better able to infer the data used to generate the model (R^2 of 0.95) than it is able to predict the data that it has not seen (an R^2 of -33080297). Let's see if I can improve this through standardization and Lasso regularization.

# Lasso to eliminate unnecessary variables

```python
lasso = LassoCV()
```

```python
def lasso_coefs(X_train_df_sc, y_train_df, alphas):
    coefs = []
    lasso_reg = Lasso()
    for a in alphas:
        lasso_reg.set_params(alpha=a)
        lasso_reg.fit(X_train_df_sc, y_train_df)
        coefs.append(lasso_reg.coef_)
        
    return coefs
```

```python
l_alphas = np.arange(400, 700, 0.2)
l_coefs = lasso_coefs(X_train_df_sc, y_train_df, l_alphas)
```

```python
l_alphas = np.arange(400, 700, 0.2)
lasso_model = LassoCV(alphas=l_alphas, cv=5)
lasso_model.fit(X_train_df_sc, y_train_df)
```

```python
lasso_optimal_alpha = lasso_model.alpha_
lasso_optimal_alpha
```

```python
lasso_model.score(X_train_df_sc, y_train_df)
```

```python
columns = X_train_df.columns
```

```python
lasso_coefs = np.array(list(zip(columns, lasso_model.coef_)))
```
