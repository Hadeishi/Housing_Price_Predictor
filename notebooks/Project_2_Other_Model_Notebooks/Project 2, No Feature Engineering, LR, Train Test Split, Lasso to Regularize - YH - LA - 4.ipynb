# Get Dummies:  Convert Continuous Var to Categorical (MSClass)

```python
train_df.ms_subclass = train_df.ms_subclass.astype('object')
```

```python
test_df.ms_subclass = test_df.ms_subclass.astype('object')
```

```python
cols_need_numeric = ['lot_frontage', 'mas_vnr_area', 'bsmtfin_sf_1', 'bsmtfin_sf_2',
                     'bsmt_unf_sf', 'total_bsmt_sf', 'garage_area']
```

```python
train_df[cols_need_numeric] = train_df[cols_need_numeric].apply(pd.to_numeric)
```

```python
test_df[cols_need_numeric] = test_df[cols_need_numeric].apply(pd.to_numeric)
```

```python
cols_need_dummies = train_df.select_dtypes(include='object').columns
```

```python
cols_need_dummies
```

```python
train_df_dummies = pd.get_dummies(train_df, columns=cols_need_dummies)
```

```python
test_df_dummies = pd.get_dummies(test_df, columns=cols_need_dummies)
```

```python
train_df_dummies.shape
```

```python
test_df_dummies.shape
```

```python
missing_cols = [col for col in train_df_dummies.columns if col not in test_df_dummies.columns]

for col in missing_cols:
    test_df_dummies[col] = [0]*test_df_dummies.shape[0]
test_df_dummies.drop('saleprice', 1, inplace=True)   ### Just to make sure...

extra_cols = [col for col in test_df_dummies.columns if col not in train_df_dummies.columns]

test_df_dummies.drop(extra_cols, 1, inplace=True)
```

# Linear Regression (for grins...)

```python
# Create target variable, y, and df of predictor var's, X
y = train_df_dummies.saleprice
X = train_df_dummies
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
slr = LinearRegression()
slr.fit(X_train, y_train)
```

```python
slr.score(X_train, y_train)
```

```python
slr.score(X_test, y_test)
```

Hard to interpret, other than to say that the overfitting with all those variables is too powerful for Train Test split to take care of. Let's abandon this avenue and go back to the original train_df_dummies and test_df_dummies. Let's try standardizing and running these through Lasso.

# Standardize Variables:

# Lasso to Regularize; screen out unnecessary columns

```python
lasso = LassoCV()
```

```python
lasso.fit(X_train_df_sc, y_train_df)
lasso.score(X_train_df_sc, y_train_df)
```

```python
lasso.coef_
```

A sparse array of beta coefficients left over after Lasso, although the ones that are left are pretty big, some e+02, some e+03, one even e+04.
