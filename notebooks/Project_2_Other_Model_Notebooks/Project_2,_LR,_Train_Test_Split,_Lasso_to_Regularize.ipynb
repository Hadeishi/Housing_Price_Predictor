# Engineer Variables:
So, discussing with my classmates, it seems that we are all following similar paths, trying to throw all of our variables into Lasso and count on it to decide what variables to push forward into our final model. Thus far, I have heard about two similar approaches, neither of which made models that fared better than our baseline "bad model." So, let's try a different approch:  engineer variables, look at interactions between variables and see if we can create a model that might be expected to fare better from a theoretical perspective.

Let's choose variables that are most likely to be maximally predictive:  overall_qual, gr_liv_area, 1st_fl_sf, year_built, full_bath, year_remodel/add, totrms_abvgr, and fireplaces. Let's feed these into a simple linear regression model,adding the interactions between overall_qual and gr_liv_area, between year_built and year remodeled, and between the total number of rooms above ground, the number of fireplaces, and the number of full bathrooms.

```python
train_df['overall_quXgr_liv_df'] = train_df['overall_qual'] * train_df['gr_liv_area']
```

```python
test_df['overall_quXgr_liv_df'] = test_df['overall_qual'] * test_df['gr_liv_area']
```

```python
train_df['year_builtXyear_remodel/add'] = train_df['year_built'] * train_df['year_remod/add']
```

```python
test_df['year_builtXyear_remodel/add'] = test_df['year_built'] * test_df['year_remod/add']
```

```python
train_df['totrmsXfireplacesXfullbath'] = train_df['totrms_abvgrd'] * train_df['fireplaces'] * train_df['full_bath']
```

```python
test_df['totrmsXfireplacesXfullbath'] = test_df['totrms_abvgrd'] * test_df['fireplaces'] * test_df['full_bath']
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(train_df.corr(), annot=True);
```

Engineered variables do correlate with saleprice to a considerable degree, although interestingly these interaction terms do covary with a number of other predictive variables, not just with the variables themselves. Introducing these interaction terms may open us up to problems caused by multi-collinearity.
